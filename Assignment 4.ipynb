```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_digits
import matplotlib.pyplot as plt 
from sklearn.preprocessing import StandardScaler
```

```python
digits = load_digits()
plt.show()
```

```python
scaler = StandardScaler()
```

```python
scaled = scaler.fit_transform(digits.data)
print(scaled)
print(scaled.T) #transpose array
print(np.shape(digits))
scaledT = scaled.T
```

```python
print(np.cov(scaled))
covMat = np.cov(scaledT) 
print(scaled.shape)
```

```python
eigen,vector = np.linalg.eig(covMat)
print(eigen)
print(vector)
```

```python
print(scaled.dot(vector.T[0]))
print(scaled.dot(vector.T[1]))
print(scaled.dot(vector.T[2]))
```

```python
df = pd.DataFrame()
df = df.fillna(0)
for i in range(10):
    vec = scaled.dot(vector.T[i])
    
    colName = "d" + str(i)
    df.insert(i, colName, vec, True)
print(df)
```
